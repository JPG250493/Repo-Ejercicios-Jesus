### API de GitHub

### URL Base
Para la API de GitHub v3
```
https://api.github.com
```

### Autenticación
La API de GitHub requiere autenticación para la mayoría de sus operaciones. Esto se suele hacer mediante un token de acceso personal (Personal Access Token, PAT) que se puede generar en los ajustes de tu cuenta de GitHub. El token se envía en el encabezado de autorización de tus solicitudes HTTP:

```python
headers = {"Authorization": "token TU_TOKEN_DE_ACCESO"}
```

### Endpoints Comunes
1. **Repositorios**:
   - Obtener repositorios de un usuario: `GET /users/:username/repos`
   - Crear un repositorio: `POST /user/repos`
   - Obtener detalles de un repositorio: `GET /repos/:owner/:repo`

2. **Issues**:
   - Listar issues: `GET /issues`
   - Crear un issue: `POST /repos/:owner/:repo/issues`

3. **Pull Requests**:
   - Listar pull requests: `GET /repos/:owner/:repo/pulls`
   - Crear un pull request: `POST /repos/:owner/:repo/pulls`

4. **Commits**:
   - Listar commits en un repositorio: `GET /repos/:owner/:repo/commits`

5. **Usuarios**:
   - Obtener detalles de un usuario: `GET /users/:username`


### Documentación Oficial
La [documentación oficial de la API de GitHub](https://docs.github.com/en/rest) es muy completa y proporciona detalles sobre todos los endpoints, parámetros y opciones disponibles, así como ejemplos de uso.
