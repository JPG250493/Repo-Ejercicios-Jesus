```python
import numpy as np


üì¶ Creaci√≥n de arrays

| Funci√≥n / M√©todo                | Explicaci√≥n                   | Ejemplo                         |
| ------------------------------- | ----------------------------- | ------------------------------- |
| `np.array([1,2,3])`             | Crea un array desde una lista | `np.array([1,2,3])`             |
| `np.zeros((2,3))`               | Array de ceros                | `np.zeros((2,3))`               |
| `np.ones((3,2))`                | Array de unos                 | `np.ones((3,2))`                |
| `np.arange(0,10,2)`             | Secuencia de valores          | `np.arange(0,10,2)`             |
| `np.linspace(0,1,5)`            | Valores igualmente espaciados | `np.linspace(0,1,5)`            |
| `np.eye(3)`                     | Matriz identidad              | `np.eye(3)`                     |
| `np.empty((2,2))`               | Array sin inicializar         | `np.empty((2,2))`               |
| `np.random.rand(2,3)`           | N√∫meros aleatorios [0,1)      | `np.random.rand(2,3)`           |
| `np.random.randint(0,10,(3,3))` | Enteros aleatorios            | `np.random.randint(0,10,(3,3))` |


üîç Atributos de arrays
| Atributo    | Explicaci√≥n                        | Ejemplo                |
| ----------- | ---------------------------------- | ---------------------- |
| `.shape`    | Dimensiones del array              | `a.shape`              |
| `.ndim`     | N√∫mero de dimensiones              | `a.ndim`               |
| `.size`     | N√∫mero total de elementos          | `a.size`               |
| `.dtype`    | Tipo de datos                      | `a.dtype`              |
| `.itemsize` | Tama√±o en bytes de cada elemento   | `a.itemsize`           |
| `.nbytes`   | Tama√±o total en bytes              | `a.nbytes`             |
| `.T`        | Transpuesta                        | `a.T`                  |
| `.flat`     | Iterador sobre todos los elementos | `for x in a.flat: ...` |


‚öôÔ∏è Operaciones b√°sicas
| Operaci√≥n          | Explicaci√≥n                     | Ejemplo            |
| ------------------ | ------------------------------- | ------------------ |
| `+,-,*,/`          | Operaciones elemento a elemento | `a + b`            |
| `np.add(a,b)`      | Suma                            | `np.add(a,b)`      |
| `np.subtract(a,b)` | Resta                           | `np.subtract(a,b)` |
| `np.multiply(a,b)` | Multiplicaci√≥n                  | `np.multiply(a,b)` |
| `np.divide(a,b)`   | Divisi√≥n                        | `np.divide(a,b)`   |
| `np.power(a,2)`    | Potencia                        | `np.power(a,2)`    |
| `np.sqrt(a)`       | Ra√≠z cuadrada                   | `np.sqrt(a)`       |
| `np.exp(a)`        | Exponencial                     | `np.exp(a)`        |
| `np.log(a)`        | Logaritmo natural               | `np.log(a)`        |
| `np.abs(a)`        | Valor absoluto                  | `np.abs(a)`        |
| `np.round(a,2)`    | Redondeo                        | `np.round(a,2)`    |


üìä Funciones estad√≠sticas
| Funci√≥n                         | Explicaci√≥n              | Ejemplo               |
| ------------------------------- | ------------------------ | --------------------- |
| `np.sum(a)`                     | Suma total               | `np.sum(a)`           |
| `np.mean(a)`                    | Media                    | `np.mean(a)`          |
| `np.median(a)`                  | Mediana                  | `np.median(a)`        |
| `np.std(a)`                     | Desviaci√≥n est√°ndar      | `np.std(a)`           |
| `np.var(a)`                     | Varianza                 | `np.var(a)`           |
| `np.min(a)` / `np.max(a)`       | M√≠nimo / M√°ximo          | `np.min(a)`           |
| `np.argmin(a)` / `np.argmax(a)` | √çndice del m√≠nimo/m√°ximo | `np.argmax(a)`        |
| `np.cumsum(a)`                  | Suma acumulada           | `np.cumsum(a)`        |
| `np.cumprod(a)`                 | Producto acumulado       | `np.cumprod(a)`       |
| `np.percentile(a,50)`           | Percentil                | `np.percentile(a,50)` |


üîÑ Indexaci√≥n y slicing
| Operaci√≥n    | Explicaci√≥n         | Ejemplo      |
| ------------ | ------------------- | ------------ |
| `a[0]`       | Primer elemento     | `a[0]`       |
| `a[-1]`      | √öltimo elemento     | `a[-1]`      |
| `a[1:4]`     | Slicing             | `a[1:4]`     |
| `a[::2]`     | Saltos              | `a[::2]`     |
| `a[mask]`    | M√°scara booleana    | `a[a>3]`     |
| `a[[0,2,4]]` | √çndices espec√≠ficos | `a[[0,2,4]]` |
| `a.T`        | Transpuesta de 2D   | `a.T`        |


üß© Manipulaci√≥n de arrays
| Funci√≥n                                 | Explicaci√≥n                | Ejemplo                   |
| --------------------------------------- | -------------------------- | ------------------------- |
| `a.reshape(2,3)`                        | Cambiar forma              | `a.reshape(2,3)`          |
| `a.flatten()`                           | Aplanar array              | `a.flatten()`             |
| `a.ravel()`                             | Aplanar sin copia          | `a.ravel()`               |
| `np.concatenate([a,b], axis=0)`         | Concatenar arrays          | `np.concatenate([a,b])`   |
| `np.stack([a,b], axis=0)`               | Apilar arrays              | `np.stack([a,b], axis=0)` |
| `np.hstack([a,b])` / `np.vstack([a,b])` | Apilar horizontal/vertical | `np.hstack([a,b])`        |
| `np.split(a,3)`                         | Dividir array              | `np.split(a,3)`           |
| `np.tile(a,2)`                          | Repetir array              | `np.tile(a,2)`            |


üî¢ √Ålgebra lineal
| Funci√≥n            | Explicaci√≥n                  | Ejemplo            |
| ------------------ | ---------------------------- | ------------------ |
| `np.dot(a,b)`      | Producto punto               | `np.dot(a,b)`      |
| `np.matmul(a,b)`   | Multiplicaci√≥n de matrices   | `np.matmul(a,b)`   |
| `np.linalg.inv(a)` | Inversa                      | `np.linalg.inv(a)` |
| `np.linalg.det(a)` | Determinante                 | `np.linalg.det(a)` |
| `np.linalg.eig(a)` | Eigenvalores y eigenvectores | `np.linalg.eig(a)` |
| `np.transpose(a)`  | Transpuesta                  | `np.transpose(a)`  |


üîÑ Comparaciones y m√°scaras
| Operaci√≥n                  | Explicaci√≥n                   | Ejemplo                   |
| -------------------------- | ----------------------------- | ------------------------- |
| `a > 5`                    | Mayor que                     | `a > 5`                   |
| `a < 3`                    | Menor que                     | `a < 3`                   |
| `a == 2`                   | Igual                         | `a == 2`                  |
| `np.logical_and(a>1, a<5)` | AND l√≥gico                    | `np.logical_and(a>1,a<5)` |
| `np.logical_or(a<2, a>5)`  | OR l√≥gico                     | `np.logical_or(a<2,a>5)`  |
| `np.where(a>5)`            | √çndices que cumplen condici√≥n | `np.where(a>5)`           |
